```python
avg_crim=data['CRIM'].mean()
data['CRIM'].replace(np.nan,avg_crim,inplace=True)

avg_zn=data['ZN'].mean()
data['ZN'].replace(np.nan,avg_zn,inplace=True)

avg_indus=data['INDUS'].mean()
data['INDUS'].replace(np.nan,avg_indus,inplace=True)

avg_chas=data['CHAS'].mean()
data['CHAS'].replace(np.nan,avg_chas,inplace=True)

avg_age=data['AGE'].mean()
data['AGE'].replace(np.nan,avg_age,inplace=True)

avg_lstat=data['LSTAT'].mean()
data['LSTAT'].replace(np.nan,avg_lstat,inplace=True)
```

```python
X=data.iloc[:,0:13]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

model=make_pipeline(StandardScaler(with_mean=False),LinearRegression())
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse=mean_squared_error(y_test,y_pred)
rmse=mean_squared_error(y_test,y_pred,squared=False)
r2=r2_score(y_test,y_pred)

print("MSE:",mse)
print("RMSE:",rmse)
print("R squared:",r2)
```
