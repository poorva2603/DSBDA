```python
#mean of each column
print("mean of First year:   Sem 1-> ",data['First year:   Sem 1'].mean())
print("mean of First year:   Sem 2-> ",data['First year:   Sem 2'].mean())
print("mean of Second year:   Sem 1-> ",data['Second year:   Sem 1'].mean())
print("mean of Second year:   Sem 2-> ",data['Second year:   Sem 2'].mean())
```

```python
#median
print("median of First year:   Sem 1-> ",data['First year:   Sem 1'].median())
print("median of First year:   Sem 2-> ",data['First year:   Sem 2'].median())
print("median of Second year:   Sem 1-> ",data['Second year:   Sem 1'].median())
print("median of Second year:   Sem 2-> ",data['Second year:   Sem 2'].median())
```

```python
#mode
print("mode of First year:   Sem 2-> ",data['First year:   Sem 1'].mode())
print("mode of Second year:   Sem 1-> ",data['First year:   Sem 2'].mode())
print("mode of Second year:   Sem 2-> ",data['Second year:   Sem 1'].mode())
print("mode of Second year:   Sem 2-> ",data['Second year:   Sem 2'].mode())
```

```python
#min value
print("min value:",data['First year:   Sem 1'].min())
#max value
print("max value:",data['First year:   Sem 1'].max())
#standard deviation
print("standard deviation:",data['First year:   Sem 1'].std())
#variance
print("variance:",data['First year:   Sem 1'].var())
```
