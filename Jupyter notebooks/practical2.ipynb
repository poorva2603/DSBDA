```python
#Scanning outliers
print(sns.boxplot(y=data['First year:   Sem 1']))
```

```python
print(sns.boxplot(y=data['First year:   Sem 2']))
```

```python
print(sns.boxplot(y=data['Second year:   Sem 1']))
```

```python
print(sns.boxplot(y=data['Second year:   Sem 2']))
```

```python
#removing outliers
removed_outliers=data[data['Second year:   Sem 1']>=8.0]
print(sns.boxplot(removed_outliers['Second year:   Sem 1']))
```

```python
#removing outliers
removed_outliers=data[data['Second year:   Sem 2']>=8.0]
print(sns.boxplot(removed_outliers['Second year:   Sem 2']))
```

```python
removed_outliers_indices = np.where((data['First year:   Sem 1'] > 0) & (data['First year:   Sem 1'] < 80))
filtered_data = data.iloc[removed_outliers_indices]

plt.figure(figsize=(6,4))
sns.boxplot(y=filtered_data['First year:   Sem 1'])
plt.show()
```

```python
removed_outliers_indices=np.where((data['First year:   Sem 2']>0)&(data['First year:   Sem 2']<80))
filtered_data=data.iloc[removed_outliers_indices]
plt.figure(figsize=(6,4))
sns.boxplot(y=filtered_data['First year:   Sem 2'])
plt.show()
```
